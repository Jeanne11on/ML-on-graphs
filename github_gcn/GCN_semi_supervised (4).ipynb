# Preprocessing

```python
# Create adjacency matrix 

import csv

# Replace with the path to your CSV file
csv_path = "/content/git_web_ml/musae_git_edges.csv"

# Read the edges from the CSV file
edges = []
with open(csv_path) as f:
    reader = csv.reader(f)
    next(reader) #skip the first line so we avoid the headers
    for row in reader:
        edges.append((int(row[0]), int(row[1])))

# Determine the number of nodes
N = max(max(e) for e in edges) + 1

# Create an N by N matrix filled with zeros
adj_matrix = [[0] * N for _ in range(N)]

# Set the entries in the adjacency matrix
for i, j in edges:
    adj_matrix[i][j] = 1
    adj_matrix[j][i] = 1

# Print the adjacency matrix
for row in adj_matrix:
    print(row)
```
